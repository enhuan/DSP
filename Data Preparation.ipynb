## **Checking Missing Values**

### **Customer Data**

```python
# Check missing values in customer data
customer.isnull().sum()
```

```python
# Check if missing values for full_name, gender, age, income, nric match
individual_null_cols = ['full_name', 'gender', 'age', 'income', 'nric']
all_individual_nulls_match = True

for i in range(len(individual_null_cols) - 1):
    if not (customer[individual_null_cols[i]].isnull() == customer[individual_null_cols[i+1]].isnull()).all():
        all_individual_nulls_match = False
        break

if all_individual_nulls_match:
    print("Yes")
else:
    print("No")
```

```python
# Check if missing values for company_name, industry_sector, sme_flag match
business_null_cols = ['company_name', 'industry_sector', 'sme_flag']
all_business_nulls_match = True

for i in range(len(business_null_cols) - 1):
    if not (customer[business_null_cols[i]].isnull() == customer[business_null_cols[i+1]].isnull()).all():
        all_business_nulls_match = False
        break

if all_business_nulls_match:
    print("Yes")
else:
    print("No")
```

```python
# Check if all rows with missing values for full_name are business companies
if (customer[customer['full_name'].isnull()]['customer_type'] == 'Business').all():
    print("Yes")
else:
    print("No")
```

```python
# Check if all rows with missing values for company_name are individuals
if (customer[customer['company_name'].isnull()]['customer_type'] == 'Individual').all():
    print("Yes")
else:
    print("No")
```

```python
# Fill missing values with NA except for age and subscription_end_date
for col in customer.columns:
    if col != 'age' and col != 'subscription_end_date':
        customer[col] = customer[col].fillna('NA')
    elif col != 'age' and col != 'subscription_start_date':
        pass
```

# **Exploratory Data Analysis (EDA)**

### **Customer Data**

```python
# Boxplot of Age
plt.figure(figsize=(8, 6))
sns.boxplot(x=customer['age'], color='purple')
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.ylabel('')
plt.show()
```

```python
# Visualize categorical columns using bar charts
customer_categorical_cols = ['customer_type',
                             'industry_sector',
                             'state',
                             'subscription_plan',
                             'status',
                             'payment_method',
                             'billing_cycle',
                             'preferred_language',
                             'gender',
                             'income',
                             'sme_flag',
                             'kyc_status']

for col in customer_categorical_cols:
    # Filter out 'NA' values
    filtered_data = customer[customer[col] != 'NA']

    plt.figure(figsize=(10, 6))
    sns.countplot(y=filtered_data[col], order=filtered_data[col].value_counts().index, palette='plasma', hue=filtered_data[col], legend=False)
    plt.title(f'Distribution of {col.replace("_", " ").title()}')
    plt.xlabel('Count')
    plt.ylabel(col.replace("_", " ").title())
    plt.tight_layout()
    plt.show()
```

### **Billing Data**

```python
# Boxplots of subtotal_amount, discount_amount, tax_amount, total_amount
billing_amount_cols = ['subtotal_amount', 'discount_amount', 'tax_amount', 'total_amount']

for col in billing_amount_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=billing[col])
    plt.title(f'Boxplot of {col.replace("_", " ").title()}')
    plt.xlabel('Amount')
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```

```python
# Visualize categorical columns using bar charts
billing_categorical_cols = ['pricing_tier', 'currency', 'payment_method']

for col in billing_categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=billing[col], order=billing[col].value_counts().index, palette='viridis', hue=billing[col], legend=False)
    plt.title(f'Distribution of {col.replace("_", " ").title()}')
    plt.xlabel('Count')
    plt.ylabel(col.replace("_", " ").title())
    plt.tight_layout()
    plt.show()
```

### **Service Data**

```python
# Boxplots of bandwidth_mbps, monthly_usage_gb, avg_latency_ms, jitter_ms, monthly_outages
service_num_cols = ['bandwidth_mbps', 'monthly_usage_gb', 'avg_latency_ms', 'jitter_ms', 'monthly_outages']

for col in service_num_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=service[col], color='grey')
    plt.title(f'Boxplot of {col.replace("_", " ").title()}')
    plt.xlabel(col.replace("_", " ").title())
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```

```python
# Visualize categorical columns using bar charts
service_categorical_cols = ['service_category', 'service_status', 'region', 'access_technology', 'router_model', 'device_os', 'sla_tier']

for col in service_categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=service[col], order=service[col].value_counts().index, palette='cividis', hue=service[col], legend=False)
    plt.title(f'Distribution of {col.replace("_", " ").title()}')
    plt.xlabel('Count')
    plt.ylabel(col.replace("_", " ").title())
    plt.tight_layout()
    plt.show()
```
